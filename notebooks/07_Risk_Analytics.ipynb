```python
import pandas as pd

from src.signals import SignalGenerator
from src.portfolio.constructor import PortfolioConstructor
from src.backtester import Backtester
from src.risk import RiskAnalytics
```

```python
prices = pd.read_csv(
    "../data/raw/daily_prices.csv",
    index_col=0,
    parse_dates=True
)
```

```python
signal_engine = SignalGenerator()

weekly_returns = signal_engine.weekly_returns(prices)

signals = signal_engine.generate_long_short_signals(
    weekly_returns
)

portfolio = PortfolioConstructor()

weights = portfolio.equal_weight_portfolio(signals)

backtester = Backtester()

strategy_returns = backtester.calculate_portfolio_returns(
    weights,
    weekly_returns
)
```

```python
risk = RiskAnalytics()
```

```python
rolling_vol = risk.rolling_volatility(
    strategy_returns
)

rolling_vol.tail()
```

```python
rolling_sharpe = risk.rolling_sharpe(
    strategy_returns
)

rolling_sharpe.tail()
```

```python
drawdown = risk.drawdown_series(
    strategy_returns
)

drawdown.tail()
```

```python
print(
    "Weekly VaR (95%):",
    risk.value_at_risk(strategy_returns)
)

print(
    "Weekly CVaR (95%):",
    risk.conditional_var(strategy_returns)
)
```
